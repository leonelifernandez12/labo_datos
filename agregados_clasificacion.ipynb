```python
from sklearn.model_selection import train_test_split

X = df_datos.drop(columns=['SECTOR'])
y = df_datos['SECTOR']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)

X_train_scaled = scaler.fit_transform(X_train)

# ---------------------------------------------------

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

valores_k_impares = [k for k in range(1, 21) if k % 2 != 0]
resultados_cv = []

for k in valores_k_impares:
    knn = KNeighborsClassifier(n_neighbors=k)
    
    puntajes = cross_val_score(knn, X_train_scaled, y_train, cv=5, scoring='accuracy')
    resultados_cv.append(puntajes.mean())

k_optimo = valores_k_impares[np.argmax(resultados_cv)]
print(f"K óptimo: {k_optimo}")

# ---------------------------------------------------

X_test_scaled = scaler.transform(X_test)
knn_con_k_optimo = KNeighborsClassifier(n_neighbors=k_optimo)

knn_con_k_optimo.fit(X_train_scaled, y_train)

y_pred = knn_con_k_optimo.predict(X_test_scaled)

precision_test = knn_con_k_optimo.score(X_test_scaled, y_test)

precision_test

#X2 = df_datos_sin_variables_elegidas.drop(columns=['SECTOR'])
y2 = df_datos_sin_variables_elegidas['SECTOR']

X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X2, y2, test_size=0.20, random_state=42
)

X_train_scaled2 = scaler.fit_transform(X_train2)

valores_k_impares2 = [k for k in range(1, 21) if k % 2 != 0]
resultados_cv2 = []

for k2 in valores_k_impares2:
    knn2 = KNeighborsClassifier(n_neighbors=k2)
    puntajes2 = cross_val_score(knn2, X_train_scaled2, y_train2, cv=5, scoring='accuracy')
    resultados_cv2.append(puntajes.mean())

k_optimo2 = valores_k_impares2[np.argmax(resultados_cv2)]
print(f"K óptimo: {k_optimo2}")

X_test_scaled2 = scaler.transform(X_test2)
knn_con_k_optimo2 = KNeighborsClassifier(n_neighbors=k_optimo2)
knn_con_k_optimo2.fit(X_train_scaled2, y_train2)

y_pred2 = knn_con_k_optimo2.predict(X_test_scaled2)
precision_test2 = knn_con_k_optimo2.score(X_test_scaled2, y_test2)

print(precision_test2)

#K óptimo: 19
#Scoring: 0.5357255272269437
```
